Задача. 
Составить действующую модель охоты кота на мышей. 

Ввод данных:

1) сторона квадрата

2) радиус действия кота 

3) кол-во мышей

Если расстояние от центра кота до центра мыши <= радиусу охоты кота, то кот съедает мышь и на место. Если больше, то кот сидит, мышь убегает.

Графическое иллюстрирование, развитый и многоязычный интерфейс. 



# Русский язык

# 1) Запуск приложения, интерфейс, ввод данных и получение результата:

## Для того чтобы работать с приложением, следует скачать с Google Disk папку catAndMice_Release, распаковать архив в удобное место, открыть папку bin и запустить catAndMouse.exe.

## Язык по умолчанию - Русский. Для того чтобы изменить язык, следует выбрать из выпадающего списка, расположенного на панеле инструментов, другой язык. Каждый элемент выпадающего списка имеет иконку-флаг и название языка. Язык можно изменять в любой момент времени работы приложения.

## В течение работы приложения звучит саундтрек.

## При клике мышью по кнопке "Начать игру!" (надпись зависит от выбранного языка), расположенной в нижней части виджета, появляются спинбоксы, то есть виджеты, которые позволяют выбирать из фиксированного набора значений.

## Первое, что необходимо ввести - это длину стороны квадрата. Квадрат - область, в которой появляются мыши в разных местах. Сторона квадрата может быть от 300-500 ед. Пользователь может вводить только числа в SpinBox, если введено значение меньше 300 ед., то автоматически длина стороны квадрата равна 300 ед, то есть минимальному значению (при повторной игре равна прошлому значению). Число, которое больше максимального значения, пользователь ввести не может.

## Число в SpinBox можно изменять с помощью колёсика мыши (фокус на Spinbox), клавиатуры, стрелочек. Далее следует кликнуть по кнопке "OK".

## Второе, что необходимо ввести - радиус охоты кота. Радиус может быть равен нулю, это минимальное значение, а максимальное значение - радиус описанной окружности вокруг квадрата, на случай, если кот охотится по всей области проживания мышей. Далее следует кликнуть по кнопке "OK".

## Третье, что нужно указать - количество мышей. Полоса прокрутки ограничивает количество мышей от 1 до 50 штук. Далее следует кликнуть по кнопке "OK".

## Снизу виджета указаны текущие значения параметров.

## Когда всё поля заполнены, появляется прогресс-бар, начинается обработка данных и установка сцены.

## Как только значение прогресс-бара равно 100 процентам, появляется сцена с котом и мышами.

## На сцене квадрат с чёрным контуром показывает область проживания мышей, а окружность с зелёным контуром - радиус охоты кота. За окружность кот не выходит. Если мышь появилась за пределами окружности, она тут же убегает, кот сидит на месте. Если мышь появилась внутри окружности, кот её атакует и возвращается на место. Как только всё мыши побывали на сцене, происходит подсчёт результата. Выключается основная мелодия, запускается звук "награждения".

##Если съеденные мыши > спасшиеся мыши, то кот получает 3 звезды.
##Если съеденные мыши == спасшиеся мыши, то кот получает 2 звезды.
##Если съеденные мыши < спасшиеся мыши, то кот получает 1 звезду.

##Под звёздами показано количество мышей каждой категории.
##Снизу виджета находится кнопка "Заново", кликнув по которой, можно начать игру заново. При нажатии на крестик, расположенный в правом верхнем углу, приложение завершит работу.

# English

# 1) Launching the application, interface, entering data and getting the result:

## In order to work with the application, download the catAndMice_Release folder from Google Disk, unzip the archive to a convenient location, open the bin folder and run catAndMouse.exe.

## The default language is Russian. In order to change the language, select another language from the drop-down list located on the toolbar. Each item in the dropdown list has a flag icon and a language name. The language can be changed at any time while the application is running.

## The soundtrack plays while the application is running.

## When clicking on the button "Start the game!" (the inscription depends on the selected language), located at the bottom of the widget, spinboxes appear, that is, widgets that allow you to select from a fixed set of values.

## The first thing to enter is the length of the side of the square. The square is an area where mice spawn in various places. The side of the square can be from 300-500 units. The user can only enter numbers in the SpinBox, if the value is less than 300 units, then automatically the length of the side of the square is 300 units, that is, the minimum value. A number greater than the maximum value cannot be entered by the user.

## The number in the SpinBox can be changed using the mouse wheel (focus on the Spinbox), keyboard, arrows. Next, click on the "OK" button.

## The second thing to enter is the cat's hunting radius. The radius can be zero, this is the minimum value, and the maximum value is the radius of the circumscribed circle around the square, in case the cat hunts throughout the area where mice live. Next, click on the "OK" button.

## The third thing to specify is the number of mice. The scrollbar limits the number of mice from 1 to 50. Next, click on the "OK" button.

## At the bottom of the widget are the current parameter values.

## When all fields are filled, a progress bar appears, data processing and scene installation begin.

## On the stage, a square with a black outline shows the area where mice live, and a circle with a green outline shows the radius of the cat's hunting. The cat does not go beyond the circle. If the mouse appeared outside the circle, they immediately run away, the cat sits still. If the mouse appears inside the circle, the cat attacks it and returns to its place. As soon as all the mice have been on the stage, the result is calculated. The main melody turns off, the "rewarding" sound starts.

##If eaten mice > escaped mice, then the cat gets 3 stars.
##If eaten mice == escaped mice, then the cat gets 2 stars.
##If eaten mice < escaped mice, then the cat gets 1 star.

##Under the stars shows the number of mice in each category.
##At the bottom of the widget is the "Restart" button, by clicking on which you can start the game again. When you click on the cross located in the upper right corner, the application will exit.
